```python
# Score Customer by RFM Model with a PercentileRanking Logic
hecuba['Monetary_Percentile'] = hecuba['revenue'].rank(pct=True)*100
hecuba['Recency_Percentile'] = hecuba['days_elapsed'].rank(pct=True)*100
hecuba['Frequency_Percentile'] = hecuba['transactions'].rank(pct=True)*100
hecuba
```

```python
hecuba.describe()
```

```python
# Define Function for Score Monetary PercentileRank
def monetary_score(monetary):
    if 0 <= monetary <= 25:
        return 1
    elif 25 < monetary <= 50:
        return 2
    elif 50 < monetary <= 75:
        return 3
    elif 75 < monetary <= 100:
        return 4

hecuba['Monetary_Score'] = hecuba['Monetary_Percentile'].apply(monetary_score)
hecuba
```

```python
# Define Function for Score Recency PercentileRank
def recency_score(recency):
    if 0 <= recency <= 25:
        return 4
    elif 25 < recency <= 50:
        return 3
    elif 50 < recency <= 75:
        return 2
    elif 75 < recency <= 100:
        return 1

hecuba['Recency_Score'] = hecuba['Recency_Percentile'].apply(recency_score)
hecuba
```

```python
# Define Function for Score Frequency PercentileRank
def frequency_score(frequency):
    if 0 <= frequency <= 25:
        return 1
    elif 25 < frequency <= 50:
        return 2
    elif 50 < frequency <= 75:
        return 3
    elif 75 < frequency <= 100:
        return 4


hecuba['Frequency_Score'] = hecuba['Frequency_Percentile'].apply(frequency_score)
hecuba
```

```python
# Create a column storing the Customers' scores in Recency, Frequency, and Monetary for Labeling RFM Group
hecuba['RFM_Group'] = hecuba['Recency_Score'].astype(str) + hecuba['Frequency_Score'].astype(str) + hecuba['Monetary_Score'].astype(str)
hecuba['RFM_Group']
```

```python
hecuba.info()
```

```python
# Aggregate Statistics for Exploring RFM Group Coupon Return Rate and corresponding generated revenue
hecuba_rfm = hecuba.groupby('RFM_Group')[['email_offer','redeemed','revenue']].sum().reset_index()
hecuba_rfm['conv_rate'] = hecuba_rfm['redeemed']/hecuba_rfm['email_offer'] * 100
hecuba_rfm
```

```python
# Calculate Targeting RFM Group share out of all RFM Group that has a Coupon Return Rate exceeding 17% 
Target_rfm_share = hecuba_rfm[hecuba_rfm['conv_rate'] > 17]['RFM_Group'].count()/hecuba_rfm['RFM_Group'].count()*100
Target_rfm_share
```

```python
# Calculate Targeting RFM Group Coupon Redemptation Share out of all Redeemed Coupons
Target_rfm_Redemptation_share = hecuba_rfm[hecuba_rfm['conv_rate']>17]['redeemed'].sum()/hecuba_rfm['redeemed'].sum()*100
Target_rfm_Redemptation_share
```

```python
# Calculate Targeting RFM Group Coupon Redemptation Rate
Target_rfm_Redemptation = hecuba_rfm[hecuba_rfm['conv_rate']>17]['redeemed'].sum()/hecuba_rfm[hecuba_rfm['conv_rate']>17]['email_offer'].sum()*100
Target_rfm_Redemptation
```

```python
# Based on 70 more available coupons to be used, calculate total additional email offers needed to be sent
Additional_Email_Offer = round(70/Target_rfm_Redemptation * 100,0)
Additional_Email_Offer
```

```python
# Calculate current total email offers sent to the Targetting RFM Group
Current_Email_Offer = hecuba_rfm[hecuba_rfm['conv_rate']>17]['email_offer'].sum()
Current_Email_Offer
```

```python
# Additional Email_Offer increase trend
Email_Offer_Increase = Additional_Email_Offer/Current_Email_Offer*100
Email_Offer_Increase
```

```python
# Prepare data for calculating Average Revenue Per Customer in Targetting RFM Group
hecuba_rfm_rev = hecuba.groupby('RFM_Group')[['email_offer','revenue']].agg({
    'email_offer': 'count',
    'revenue': 'sum'
}).reset_index().rename(columns={'email_offer':'Total_Customer'})
hecuba_rfm_rev
```

```python
# Prepare data for calculating Average Revenue Per Customer in Targetting RFM Group
filtered_group = hecuba_rfm_rev[hecuba_rfm_rev['RFM_Group'].isin(hecuba_rfm[hecuba_rfm['conv_rate']>17]['RFM_Group'])]
filtered_group
```

```python
# Calculate Average Revenue per Customer in Targetting Group
avg_rev_per_customer = filtered_group['revenue'].sum()/filtered_group['Total_Customer'].sum()
avg_rev_per_customer
```

```python
# Calculate Current Total Revenue among Targetting Group
total_rev = filtered_group['revenue'].sum()
total_rev
```

```python
# If 50% coupon redeemed by improved campaign backed by RFM Analysis
safe_projected_rev = 0.5*70*avg_rev_per_customer
safe_projected_rev
```

```python
# If 100% coupon redeemed by improved campaign backed by RFM Analysis
match_projected_rev = 70*avg_rev_per_customer
match_projected_rev
```

```python
# If 150% coupon redeemed by improved campaign backed by RFM Analysis (this will exceed the limit of available coupons for redemptation (204))
reach_projected_rev = 1.5*70*avg_rev_per_customer
reach_projected_rev
```

```python
# Prepare for final projected outcome table
net_safe = total_rev + safe_projected_rev
net_match = total_rev + match_projected_rev
net_reach = total_rev + reach_projected_rev
```

```python
Projected_Return = {
    'safe':[35,safe_projected_rev,net_safe],
    'match':[70,match_projected_rev,net_match],
    'reach':[105,reach_projected_rev,net_reach]
}
Projected_Return
```

```python
Projected_Return_ct = pd.DataFrame(Projected_Return, index=['Coupon Redemptation','Projected Revenue', 'Net Revenue'])
Projected_Return_ct
```
